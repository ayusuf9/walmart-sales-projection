```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestRegressor
```

## Loading the required datasets

```python
df_train = pd.read_csv("train.csv")
df_test = pd.read_csv("test.csv")
df_features = pd.read_csv("features.csv")
df_stores = pd.read_csv("stores.csv")
```

```python
#I will be checking for any missing values in the datasets before 
#using them for analysis and training the regression model.

dfs = {
    'df_train': df_train, 
    'df_test': df_test, 
    'df_features': df_features, 
    'df_stores': df_stores
}

for name, df in dfs.items():
    total_nas = df.isna().sum().sum()
    print(f"Total missing values in {name} across all columns is: {total_nas}")
    print()
```

```python
# Based on the above, we need to fill the missing values in df_features
# But we need details of those columns
df_features.isna().sum()
```

```python
# Based on the above the features in df_features.csv that need imputation are:
# MarkDown1, MarkDown2, MarkDown3, MarkDown4, MarkDown5, CPI and Unemployment 

df_features[['MarkDown1', 'MarkDown2','MarkDown3','MarkDown4', 
'MarkDown5']] = df_features[['MarkDown1', 'MarkDown2', 'MarkDown3', 
'MarkDown4', 'MarkDown5']].fillna(0) 

df_features.dropna(inplace=True)

df_features.isna().sum()
```

```python
df_train_merge = df_train.merge(df_features, on=['Store', 'Date'], 
                                how='inner').merge(df_stores, 
                                on=['Store'], how='inner')

df_train_merge.head()
```

```python
df_train_merge[df_train_merge['Weekly_Sales'] < 0].head(3)
```

```python
df_test_merge = df_test.merge(df_features, on=['Store', 'Date'], 
                              how='inner').merge(df_stores, 
                              on=['Store'], how='inner')

df_test_merge.head()
```

```python
# After merging the dataset from the previous step, the IsHoliday column is 
#duplicated into IsHoliday_x and IsHoliday_y. Only one of them should be retained

df_train_merge.drop(['IsHoliday_y'], axis=1,inplace=True)
df_test_merge.drop(['IsHoliday_y'], axis=1,inplace=True)
```

```python
df_train_merge.rename(columns={'IsHoliday_x':'IsHoliday'},inplace=True)
df_test_merge.rename(columns={'IsHoliday_x':'IsHoliday'},inplace=True)
```

```python
"""
In this task, extract the features relevant to weekly sales to build a 
predictive model for forecasting the weekly sales at a later stage. 
To complete this task, do the following:
"""

df_train_final = df_train_clean[['Date','Store','Dept','Type','IsHoliday', 
                                 'Weekly_Sales']].reset_index(drop=True)

df_test_final = df_test_clean[['Date','Store', 'Dept', 
                               'Type','IsHoliday']].reset_index(drop=True)
```

```python
df_train_final.head()
```

```python
df_test_final.head()
```

```python
"""
encoding the categorical variables here from the final DataFrames to a numerical format. 
This is because the model can only be fed with numeric data.
"""

df_train_final_encoded = df_train_final.copy()

df_train_final_encoded[['Type','IsHoliday']] = df_train_final[['Type', 
                        'IsHoliday']].apply(LabelEncoder().fit_transform)

df_train_final_encoded.head()
```

```python
df_test_final_encoded = df_test_final.copy()

df_test_final_encoded[['Type', 'IsHoliday']] = df_test_final[['Type', 'IsHoliday']].apply(LabelEncoder().fit_transform)

df_test_final_encoded.head()
```

```python
"""
I will engineer new features from the existing columns of the final encoded 
training and test DataFrames. The required new features are as follows:

Training set: Month from the Date column and logarithm base 10 
weekly sales from the Weekly_Sales column

Test set: Month from the Date column

"""

df_train_final_encoded["Date"] = pd.to_datetime(df_train_final_encoded["Date"])

df_train_final_encoded["Month"] = df_train_final_encoded["Date"].dt.month

df_train_final_encoded['Weekly_Sales_Log10'] =  np.log10(df_train_final_encoded['Weekly_Sales'])

df_train_final_encoded.drop(['Date', 'Weekly_Sales'], axis=1, inplace=True)

df_train_final_encoded.head()
```

```python
df_test_final_encoded["Date"] = pd.to_datetime(df_test_final_encoded["Date"])

df_test_final_encoded["Month"] = df_test_final_encoded["Date"].dt.month

df_test_final_encoded.drop(['Date'],axis=1, inplace=True)

df_test_final_encoded.head()
```

```python
# I will now train a random forest regressor model to predict the weekly sales.

train_y = df_train_final_encoded['Weekly_Sales_Log10'].to_numpy()

train_x =  df_train_final_encoded[['Store', 'Dept', 'Type', 'IsHoliday', 'Month']].to_numpy()

test_x = df_test_final_encoded.to_numpy()

regressor = RandomForestRegressor()
regressor.fit(train_x, train_y)
```

```python
# I will now use the trained model to forecast weekly sales based on the test set.

predictions = regressor.predict(test_x)
df_test_final['Predictions']= 10 ** (pd.Series(predictions))
df_test_final
```

```python
# creating a line chart to visualize the models sale projections obtained

df_train_final["Date"] = pd.to_datetime(df_train_final["Date"])
df_test_final["Date"] = pd.to_datetime(df_test_final["Date"])

df_train_final.set_index('Date', inplace=True)
df_test_final.set_index('Date', inplace=True)

fig, ax = plt.subplots(1,1,figsize=(13,5))
sns.lineplot(df_train_final["Weekly_Sales"], label="Actual",errorbar=None)
sns.lineplot(df_test_final["Predictions"],label="Prediction", errorbar=None)
plt.show()
```
